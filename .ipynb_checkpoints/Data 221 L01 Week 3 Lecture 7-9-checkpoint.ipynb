```python
from bs4 import BeautifulSoup
import requests

who_covid_page_html = requests.get("http://covid19.who.int/").text
parsed_html_document = BeautifulSoup(who_covid_page_html,
                                     "html5lib") #html.parser also works
print(parsed_html_document)
```

```python
all_paragraph_elements = parsed_html_document.find_all("p")
first_paragraph_element = all_paragraph_elements[0]
second_paragraph_element = all_paragraph_elements[1]
print(first_paragraph_element)
print(second_paragraph_element)
```
